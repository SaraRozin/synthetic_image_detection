```python
import os

def view_data_tree(directory):
    """
    View Data Tree Function:
    This function takes a directory path as input and analyzes the subfolders to count the number of image files
    (.jpg, .jpeg, .png, .gif) present in each subfolder. It then prints the total number of pictures in each
    second-level subfolder and the final sum of pictures in the root directory.

    Args:
    directory (str): The root directory path to analyze.
    """
    folder_counts = {}

    # Traverse through the directory and subdirectories
    for root, dirs, files in os.walk(directory):
        # Count the number of image files in each subfolder
        pictures_count = len([file for file in files if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif'))])
        if pictures_count > 0:
            relative_path = os.path.relpath(root, start=directory)
            if '\\' in relative_path:
                second_level_folder = relative_path.split(os.sep)[0] + "\\" + relative_path.split(os.sep)[1]
                folder_counts[second_level_folder] = folder_counts.get(second_level_folder, 0) + pictures_count

    # Print the total number of pictures in each second-level subfolder
    for folder, count in folder_counts.items():
        print(f"Folder: {folder} - Total Number of Pictures in Subfolders: {count}")

    # Print the final sum of pictures in the root directory
    print(f"Final Sum of Pictures in {os.path.basename(directory)}: {sum(folder_counts.values())}")
```

## Select ~10,400 random synthtic images (416 random images * 25 generatic models = 10,400) and ~10,400 random real images (1300 real images * 8 domains = 10,400).

```python
import os
import shutil
import random
import zipfile

# Define the paths
base_path = r'Path to data directory'
archive_zip_path = os.path.join(base_path, 'archive.zip')
real_images_path = os.path.join(base_path, 'real_images')
synthetic_images_path = os.path.join(base_path, 'synthetic_images')

# Create real_images and synthetic_images folders
os.makedirs(real_images_path, exist_ok=True)
os.makedirs(synthetic_images_path, exist_ok=True)

# Define the lists of generative models and real data sets
generative_models = ['big_gan', 'cips', 'cycle_gan', 'ddpm', 'denoising_diffusion_gan', 'diffusion_gan', 'face_synthetics',
                    'gansformer', 'gau_gan', 'generative_inpainting', 'glide', 'lama', 'latent_diffusion', 'mat', 'palette',
                    'projected_gan', 'pro_gan', 'sfhq', 'stable_diffusion', 'star_gan', 'stylegan1', 'stylegan2', 'stylegan3',
                    'taming_transformer', 'vq_diffusion']
real_data_sets = ['afhq', 'celebahq', 'coco', 'ffhq', 'imagenet', 'landscape', 'lsun', 'metfaces']

def select_random_images_subset(zip_ref, folder_name, dest_folder, num_images=None, percentage_images=None):
    """
    Copy Images in Subfolders Function:
    This function extracts and copies a specified number of image files from a specific subfolder within
    a ZIP archive to a destination folder.

    Args:
    zip_ref (ZipFile): The ZipFile object.
    folder_name (str): Name of the subfolder in the ZIP archive.
    dest_folder (str): Destination folder to copy the images.
    num_images (int): Number of images to copy.
    """
    image_files = [info.filename for info in zip_ref.infolist() if info.filename.startswith(folder_name) and info.filename.lower().endswith(('.png', '.jpg', '.jpeg'))]
    if num_images:
      selected_images = random.sample(image_files, min(num_images, len(image_files)))
    else:
      selected_images = random.sample(image_files, int((percentage_images / 100 ) * len(image_files)))

    os.makedirs(dest_folder, exist_ok=True)
    for img in selected_images:
        with zip_ref.open(img) as source, open(os.path.join(dest_folder, os.path.basename(img)), 'wb') as target:
            shutil.copyfileobj(source, target)

# Open the archive.zip file and copy images from all subfolders
random_images_per_generative_model = 416
real_images_per_domain = 1300
with zipfile.ZipFile(archive_zip_path, 'r') as zip_ref:
    for model_name in generative_models:
        synthetic_dest_folder = os.path.join(synthetic_images_path, model_name)
        select_random_images_subset(zip_ref, model_name, synthetic_dest_folder, random_images_per_generative_model)
    for model_name in real_data_sets:
        real_dest_folder = os.path.join(real_images_path, model_name)
        select_random_images_subset(zip_ref, model_name, real_dest_folder, real_images_per_domain)
```

# **Handle small data**
**~ 20,000 pictures**

```python
import os
import shutil
import random

def split_data_recursively(
    data_folder, train_data_folder, test_data_folder, split_folders=True, split_images=True
):
  """
  Splits data (folders and/or images) into training and testing sets recursively.

  Args:
    data_folder (str): Path to the root directory containing the data to split.
    train_data_folder (str): Path to the destination directory for training data.
    test_data_folder (str): Path to the destination directory for testing data.
    split_folders (bool, optional): True to split folders, False to only split images within the top-level 'data' folder. Defaults to True.
    split_images (bool, optional): True to split images within folders, False to only split folders. Defaults to True.
  """

  for root, dirs, files in os.walk(data_folder):
    # Collect subdirectories to split (except excluded ones)
    subdirectories_to_split = []
    if split_folders and dirs:
      current_dir = os.path.relpath(root, data_folder)
      for subfolder in dirs:
        subdirectories_to_split.append(os.path.join(current_dir, subfolder))

    # Handle folder splitting
    if split_folders and subdirectories_to_split:
      random.shuffle(subdirectories_to_split)
      split_index = int(0.8 * len(subdirectories_to_split))  # 80/20 split for folders

      # Create training folders
      train_subdirectories = subdirectories_to_split[:split_index]
      if root == data_folder:
        train_subdirectories = subdirectories_to_split
      for subdirectory in train_subdirectories:
        os.makedirs(os.path.join(train_data_folder, subdirectory), exist_ok=True)

      # Create testing folders (using all subdirectories)
      for subdirectory in subdirectories_to_split:
        os.makedirs(os.path.join(test_data_folder, subdirectory), exist_ok=True)

    # Handle image splitting
    if split_images and files:
      # Calculate training/testing image split
      num_images = len(files)
      num_train_images = int(num_images * 0.8)  # 80/20 split for images

      # Randomly shuffle images and split
      random.shuffle(files)
      train_images = files[:num_train_images]
      test_images = files[num_train_images:]

      # Get relative path within 'data' folder
      current_relpath = os.path.relpath(root, data_folder)

      # Copy images to respective folders based on directory existence
      if os.path.exists(os.path.join(train_data_folder, current_relpath)):
        print('save 20% images to dir:', os.path.join(train_data_folder, current_relpath), 'for testing')
        for image in train_images:
          shutil.copy(os.path.join(root, image), os.path.join(train_data_folder, current_relpath, image))
      else:
        # Use all images for testing if the training directory doesn't exist
        print('save the entire dir:', os.path.join(train_data_folder, current_relpath), 'for testing')
        test_images = files
      for image in test_images:
        shutil.copy(os.path.join(root, image), os.path.join(test_data_folder, current_relpath, image))
```

```python
import os
import shutil
import random

def split_large_data_recursively(
    data_folder, train_data_folder, validation_data_folder, test_data_folder, split_folders=True, split_images=True
):
  """
  Splits data (folders and/or images) into training and testing sets recursively.

  Args:
    data_folder (str): Path to the root directory containing the data to split.
    train_data_folder (str): Path to the destination directory for training data.
    test_data_folder (str): Path to the destination directory for testing data.
    split_folders (bool, optional): True to split folders, False to only split images within the top-level 'data' folder. Defaults to True.
    split_images (bool, optional): True to split images within folders, False to only split folders. Defaults to True.
  """

  for root, dirs, files in os.walk(data_folder):
    # Collect subdirectories to split (except excluded ones)
    subdirectories_to_split = []
    if split_folders and dirs:
      current_dir = os.path.relpath(root, data_folder)
      for subfolder in dirs:
        subdirectories_to_split.append(os.path.join(current_dir, subfolder))

    # Handle folder splitting
    if split_folders and subdirectories_to_split:
      random.shuffle(subdirectories_to_split)
      split_train_index = int(0.8 * len(subdirectories_to_split))  # 80/20 split for folders
      split_validation_index = int(0.9 * len(subdirectories_to_split))
      # Create training folders
      train_subdirectories = subdirectories_to_split[:split_train_index]
      if root == data_folder:
        train_subdirectories = subdirectories_to_split
      for subdirectory in train_subdirectories:
        os.makedirs(os.path.join(train_data_folder, subdirectory), exist_ok=True)
      # Create validation folders
      validation_subdirectories = subdirectories_to_split[:split_validation_index]
      if root == data_folder:
        validation_subdirectories = subdirectories_to_split
      for subdirectory in validation_subdirectories:
        os.makedirs(os.path.join(validation_data_folder, subdirectory), exist_ok=True)
      # Create testing folders (using all subdirectories)
      for subdirectory in subdirectories_to_split:
        os.makedirs(os.path.join(test_data_folder, subdirectory), exist_ok=True)

    # Handle image splitting
    if split_images and files:
      # Calculate training/testing image split
      num_images = len(files)
      num_train_images = int(num_images * 0.8)  # 80/20 split for images
      num_validation_images = int(num_images * 0.9) - num_train_images
      # Randomly shuffle images and split
      random.shuffle(files)
      train_images = files[:num_train_images]
      validation_images = files[num_train_images: num_train_images + num_validation_images]
      test_images = files[num_train_images + num_validation_images:]

      # Get relative path within 'data' folder
      current_relpath = os.path.relpath(root, data_folder)

      # Copy images to respective folders based on directory existence
      # case when the dir in training dirs
      if os.path.exists(os.path.join(train_data_folder, current_relpath)):
        for image in train_images:
          shutil.copy(os.path.join(root, image), os.path.join(train_data_folder, current_relpath, image))

        for image in validation_images:
          shutil.copy(os.path.join(root, image), os.path.join(validation_data_folder, current_relpath, image))

        for image in test_images:
          shutil.copy(os.path.join(root, image), os.path.join(test_data_folder, current_relpath, image))

      # case when the dir not in training dirs, but in validation dirs
      elif os.path.exists(os.path.join(validation_data_folder, current_relpath)):
        for image in train_images:
          shutil.copy(os.path.join(root, image), os.path.join(validation_data_folder, current_relpath, image))

        for image in validation_images + test_images:
          shutil.copy(os.path.join(root, image), os.path.join(test_data_folder, current_relpath, image))

      # case when the dir not in training dirs, and not in validation dirs
      else:
        # Use all images for testing if the training directory doesn't exist
        print('save the entire dir:', os.path.join(train_data_folder, current_relpath), 'for testing')
        for image in files:
          shutil.copy(os.path.join(root, image), os.path.join(test_data_folder, current_relpath, image))
```
